```python
acoes = ['GOLL4.SA', 'CVCB3.SA','WEGE3.SA','MGLU3.SA','BOVA11.SA']
```

```python
acoes = ['GOLL4.SA', 'CVCB3.SA','WEGE3.SA','MGLU3.SA','BOVA11.SA']
acoes_df = pd.DataFrame()

for acao in acoes:
  # Use yfinance to download data
  data = yf.download(acao, start='2015-01-01')
  # Add the 'Close' prices to your DataFrame
  acoes_df[acao] = data['Close']
```

```python
acoes_df
```

```python
acoes_df = acoes_df.rename(columns={'GOLL4.SA': 'GOL', 'CVCB3.SA': 'CVC', 'WEGE3.SA': 'WEGE', 'MGLU3.SA': 'MGLU', 'BOVA11.SA': 'BOVA'})
```

```python
acoes_df.dropna(inplace=True)
```

```python
acoes_df_normalizado = acoes_df.copy()
for i in acoes_df_normalizado.columns[1:]:
    # Indent the line within the for loop
    acoes_df_normalizado[i] = acoes_df_normalizado[i] / acoes_df_normalizado[i][0]
```

```python
acoes_df_normalizado
```

```python
acoes_df_normalizado.plot(x= 'Date', figsize=(15,7), title = 'Histórico de açoes - normalizado');
```
