# **Plotando os principais gráficos**

Nesta prática, vamos usar as bibliotecas **Matplotlib** e **Seaborn**, dedicadas à geração de gráficos em Python, para entender como criar os principais tipos de gráficos que irão nos auxiliar em tarefas quantitativas.

## **Importando as bibliotecas**

Para começar, como de praxe, vamos importar as bibliotecas que usaremos neste notebook:

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Gerando os gráficos**

A partir daqui, vamos gerar os principais tipos de gráficos.

### **Grafico de barras**

Gráficos de barras são um dos tipos mais simples e comuns de gráficos. Eles são úteis porque permite uma visualização e comparação fácil entre quantidades diferentes.

```python
# Gerando valores demonstrativos
np.random.seed(42)
x = np.arange(14)
y = np.arange(14)
```

```python
# Plotando gráfico de barras com Matplotlib
plt.bar(x, y)
plt.title("Exemplo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

```python
# Plotando gráfico de barras com Seaborn
tips = sns.load_dataset("tips")
sns.barplot(data=tips);
```

Gráfico de coluna é uma variação do gráfico de barra, que tem a mesma função. É útil quando há muitas categorias de dados.

Gráficos de barras e de colunas podem ter variações, como colunas empilhadas, representando valores absolutos ou percentuais. Cada um será útil em situações específicas. No entanto, pode ser mais difícil comparar valores de maneira simples nos mesmos.

### **Gráficos de linha**

Gráficos de linha permitem visualizar dados contínuos, seja em relação ao tempo ou a alguma tendência, por exemplo. Eles são úteis para transmitir a variação de algo.

```python
# Gerando valores demonstrativos
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)
```

```python
# Plotando gráfico de linha com Matplotlib
plt.plot(x, y)
plt.title("Exemplo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

```python
# Plotando gráfico de linha com Seaborn
flights = sns.load_dataset("flights")
sns.lineplot(data=flights, x="year", y="passengers");
```

### **Gráfico de setores**

Conhecido popularmente como "gráfico de pizza", o gráfico de setores ou de segmentos é usado para representar porcentagens, proporções ou partes de um todo (conjunto de dados).

```python
# Gerando valores demonstrativos
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
```

```python
# Plotando gráfico de setores com Matplotlib
plt.pie(y, labels = mylabels)
plt.title("Exemplo")

plt.show()
```

Seaborn não tem uma função padrão para criar gráficos de pizza. Mas podemos usar a sintaxe no Matplotlib e adicionar uma paleta de cores do Seaborn da seguinte maneira:

```python
# Plotando gráfico de setores com Matplotlib e paleta de cores do Seaborn
tips = sns.load_dataset("tips")
colors = sns.color_palette('pastel')[0:5]
plt.pie(tips['sex'].value_counts(), labels=tips['sex'].value_counts().keys(), colors = colors, autopct='%.0f%%');
```

### **Gráfico de dispersão**

Gráfico de dispersão ou *scatter plot* serve para examinarmos a associação entre duas medidas. Por meio dele, é possível ver o quanto uma variável está ou não correlacionada a outra.

```python
# Gerando valores demonstrativos
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
```

```python
# Gerando gráfico de dispersão com Matplotlib
plt.scatter(x, y)
plt.title('Exemplo')
plt.show()
```

```python
# Gerando gráfico de dispersão com Matplotlib
sns.set_theme()
tips = sns.load_dataset("tips")
sns.relplot(
    data=tips,
    x="total_bill", y="tip");
```

### **Histogramas**

Histogramas mostram como um determinado conjunto de dados está distribuído, isto é, quais dados são mais frequentes no conjunto.

```python
# Gerando valores demonstrativos
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
```

```python
# Plotando histograma com Matplotlib
plt.hist(a, bins = [0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```

```python
# Plotando histograma com Seaborn
penguins = sns.load_dataset("penguins")
sns.displot(penguins, x="flipper_length_mm");
```

### **Boxplot**

Boxplot ou Diagrama de Caixa é um tipo de gráfico que exibe as referências de valores mínimos e máximos, primeiro e terceiro quartil, mediana e *outliers* (valores extremos) de um conjunto de dados.

Permite entender as principais medidas estatísticas de qualquer conjunto de dados.

```python
# Gerando valores demonstrativos
np.random.seed(10)
data = np.random.normal(100, 20, 200)
```

```python
# Plotando boxplot com Matplotlib
fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)
plt.title('Exemplo')

plt.show()
```

```python
# Plotando boxplot com Seaborn
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
```
